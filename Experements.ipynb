```python
def get_data(ticker):
    if type(ticker) == str:
        ticker_tmp = str(ticker)
        start = str(datetime.date(2000, 1, 1))
        end = str(datetime.date.today())
        stock = yf.download(ticker_tmp, start=start, end=end, interval="1D")
        stock = stock.reset_index()
        return stock
    else:
        return 'None'
```
